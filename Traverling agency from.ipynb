```python
import tkinter as tk
from tkinter import *
from tkinter import messagebox

root = tk.Tk()
root.geometry('644x344')
root.configure(bg="#9A9ACD")

def getvals():
    print("Submitting form")

    print(f"{namevalue.get(), phonevalue.get(), gendervalue.get(), emergencyvalue.get(), paymentmodevalue.get(), foodservicevalue.get()} ")



    with open("records.txt", "a") as f:
        f.write(f"{namevalue.get(), phonevalue.get(), gendervalue.get(), emergencyvalue.get(), paymentmodevalue.get(), foodservicevalue.get()}\n ")


#Heading
lable = tk.Label(root, text='Welcome to Raj Traverls',font="comicsansms 13 bold",pady=15,bg="#9A9ACD").grid(row=0, column=5)


#Tkinter variable for storing entries
namevalue = StringVar()
phonevalue = StringVar()
gendervalue = StringVar()
emergencyvalue = StringVar()
placeentry = StringVar()
paymentmodevalue = StringVar()
foodservicevalue = IntVar()


#Tkiter variables labels
name =  tk.Label(root,text='Name',bg="#9A9ACD").grid(row=1, column=2)
nameentry = tk.Entry(root, textvariable=namevalue).grid(row=1,column=3)

phone = tk.Label(root,text='Phone',bg="#9A9ACD").grid(row=2, column=2)
phoneentry  = tk.Entry(root, textvariable=phonevalue).grid(row=2,column=3)

gender = tk.Label(root, text="Gender",bg="#9A9ACD").grid(row=3, column=2)
genderentry = tk.Entry(root, textvariable=gendervalue).grid(row=3,column=3)

emergency  = Label(root, text="Emergency Contact",bg="#9A9ACD").grid(row=4, column=2)
emergencyentry  = tk.Entry(root, textvariable=emergencyvalue).grid(row=4, column=3)

place  = Label(root, text="Place Name",bg="#9A9ACD").grid(row=5, column=2)
placeentry  = tk.Entry(root, textvariable=placeentry).grid(row=5, column=3)


paymentmode  = tk.Label(root, text="Payment Mode",bg="#9A9ACD").grid(row=6, column=2)
paymentmodeentry  = tk.Entry(root, textvariable=paymentmodevalue).grid(row=6,column=3)

#Checkbox & Packing it
foodservice = tk.Checkbutton(text='Want to prebook your meals?',variable = foodservicevalue,bg="#9A9ACD").grid(row=7, column=3)


#Button & packing it and assigning it a command
tk.Button(text='Submit to Raj Travels',command=getvals,bg="#9A9ACD").grid(row=8, column=3)


root.mainloop()
```
